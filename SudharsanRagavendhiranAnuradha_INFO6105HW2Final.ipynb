```python
np.random.seed(42)
df3 = pd.DataFrame(np.random.randint(1,100,size=(10, 4)), columns=list('pqrs'), index=list('abcdefghij'))
df3
```

```python
# (2) Create a new column such that, each row of this df contains the row number of 
#nearest row-record by Euclidean distance. [8 points]
```

```python
nearestRow = []
minValList = []

for i, row in df3.iterrows():
    curr = row
    rest = df3.drop(i)
    distance = {}  
    for j, contestant in rest.iterrows():
        distance.update({j: round(np.linalg.norm(curr.values - contestant.values))})
    nearestRow.append(max(distance, key=distance.get))
    minValList.append(max(distance.values()))

df3['NearestRow'] = nearestRow
df3
```

```python
# (3) Create a generic function to interchange arbitrary two columns [6 points].
```

```python
def swapColumns(df, column1, column2):
    columnList = list(df.columns)
    i = columnList.index(column1)
    j = columnList.index(column2)
    columnList[j] = columnList[i]
    columnList[i] = columnList[j]
    return df[columnList]
swapColumns(df4,'c','a')
```
